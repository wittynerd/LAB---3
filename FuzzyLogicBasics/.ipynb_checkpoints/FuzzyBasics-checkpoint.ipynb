```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Membership(x,u,s):
    
    #Gaussian 
    
    return np.exp( -np.square(x-u) / (2 * np.square(s) ))
    
    pass
```

```python
x = np.linspace(0,100,1000)

u1 = 0
u2 = 50
u3 = 100
s = 15

y1 = Membership(x,u1,s)
y2 = Membership(x,u2,s)
y3 = Membership(x,u3,s)
```

```python
plt.fill_betweenx(y1,x, label = 'Low Temperature')

plt.fill(x,y2,'y', label = 'Medium Temperature')

plt.fill_betweenx(y3,x,100, color = "r", label = 'High Temperature')



plt.legend(loc = 1 , bbox_to_anchor = (1.5,1))
plt.xlabel('X : Temperature')
plt.ylabel('u(x) : Membership')
plt.grid()
```

```python
#Union of Fuzzy sets

def Union(y1,y2):
    return [ max(y1[i] , y2[i]) for i in range(len(y1))]
    
y = Union(y1,y2)
plt.plot(x,y1, label = 'Low Temperature')
plt.plot(x,y2, label = 'Medium Temperature')
plt.fill_between(x,y, color = "m" , label = 'Low OR Medium')
plt.legend(loc = 1 , bbox_to_anchor = (1.5,1))
plt.grid()
```

```python
#Intersection of Fuzzy sets

def Intersection(y1,y2):
    return [ min(y1[i] , y2[i]) for i in range(len(y1))]
     
    
y = Intersection(y1,y2)
plt.plot(x,y1, label = 'Low Temperature')
plt.plot(x,y2, label = 'Medium Temperature')
plt.fill_between(x,y, color = "b" ,label = 'Low AND Medium')
plt.legend(loc = 1 , bbox_to_anchor = (1.5,1))
plt.grid()
```

```python
# Negation of a Fuzzy Set
def Negation(y):
    return 1-y

y = Negation(y2)
plt.plot(x,y2, "g",label = 'Medium Temperature')
plt.fill(x,y, "c",label = 'NOT Medium' )
plt.legend(loc = 1 , bbox_to_anchor = (1.5,1))
plt.grid()
```

```python
#Concentration of fuzzy set
def Conc(y , k):
    return y**k

#Dilution of fuzzy set
def Dil(y , k):
    return y**(1/k)
```

```python
y_veryLow = Conc(y1,4)
y_veryHigh = Conc(y3,4)


plt.fill_betweenx(y_veryLow,x, color = 'c', label = 'Very Low Temperature')
plt.fill_betweenx(y_veryHigh,x,100, color = '#FFA500', label = 'Very Low Temperature')

plt.plot(x,y1, label = 'Low Temperature')

plt.fill(x,y2,'y', label = 'Medium Temperature')

plt.plot(x,y3 ,color = "r", label = 'High Temperature')



plt.legend(loc = 1 , bbox_to_anchor = (1.5,1))
plt.xlabel('X : Temperature')
plt.ylabel('u(x) : Membership')
plt.grid()
```
